We'll use: 
- **NumPy** for data generation and math
- **SciPy** for statistical functions
- **Matplotlib** for visualization

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

This creates 1000 random incomes with:
- **Mean** = 5000
- **Standard Deviation** = 100

```python
incomes = np.random.normal(5000, 100, 1000)
```

```python
print(f"Incomes: {np.mean(incomes)}")
```

```python
plt.hist(incomes, bins=50, alpha=0.7, edgecolor="black")
# 'plt.axvline' simply draws a vertical line at the position x across the entire y-axis of your current plot.
# Vertical line at the mean of the incomes.
plt.axvline(np.mean(incomes), color="red", linestyle="dashed", linewidth=2, label="Mean")
# Vertical line at one standard deviation above the mean.
plt.axvline(np.mean(incomes) + np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="+1 Std Dev")
# Vertical line at one standard deviation below the mean.
plt.axvline(np.mean(incomes) - np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="-1 Std Dev")
# It collects the labels and shows a little box.
plt.legend()
plt.show()
```

ðŸ‘‰ The **mean** and **median** are close, since this is a normal
distribution.\
ðŸ‘‰ The **mode** isn't very useful here, because with continuous data,
most values appear only once.

```python
print(f"Mean: {np.mean(incomes)}")
print(f"Median: {np.median(incomes)}")
print(f"Mode: {stats.mode(incomes)}")
```

```python
# This simulates adding one billionaire to the dataset.
incomes = np.append(incomes, [10000000])
```

```python
plt.hist(incomes, bins=50, alpha=0.7, edgecolor="black")
plt.axvline(np.mean(incomes), color="red", linestyle="dashed", linewidth=2, label="Mean")
plt.axvline(np.mean(incomes) + np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="+1 Std Dev")
plt.axvline(np.mean(incomes) - np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="-1 Std Dev")
plt.legend()
plt.show()
```

The **mean skyrockets**, but the **median barely changes**.\
This shows why the median is more robust to outliers.

```python
print(f"Mean: {np.mean(incomes)}")
print(f"Median: {np.median(incomes)}")
print(f"Mode: {stats.mode(incomes)}")
```

```python
# Removing the outlier
incomes = incomes[incomes != 10000000]
```

```python
print(f"Mean: {np.mean(incomes)}")
print(f"Variance: {np.var(incomes)}")
print(f"Standard Deviation: {np.std(incomes)}")
```

```python
plt.hist(incomes, bins=50, alpha=0.7, edgecolor="black")
plt.axvline(np.mean(incomes), color="red", linestyle="dashed", linewidth=2, label="Mean")
plt.axvline(np.mean(incomes) + np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="+1 Std Dev")
plt.axvline(np.mean(incomes) - np.std(incomes), color="green", linestyle="dotted", linewidth=2, label="-1 Std Dev")
plt.legend()
plt.show()
```
